```python
import numpy as np
import autokeras as ak
```

```python
import keras
```

# deneme1 2 resim giriş 2 rastgele çıkış

```python
num_instances = 10
# Generate image data.
image_data = np.random.rand(num_instances, 32, 32, 3).astype(np.float32)
# Generate numerical data.
image_data2 = np.random.rand(num_instances, 32, 32, 3).astype(np.float32)
```

```python
# Generate regression targets.
classification_target = np.random.randint(5, size=num_instances)# Generate classification labels of five classes.
classification_target2 = np.random.randint(5, size=num_instances)
```

```python
# Initialize the multi with multiple inputs and outputs.
model = ak.AutoModel(
    inputs=[ak.ImageInput(), ak.Input()],
    outputs=[
        ak.RegressionHead(metrics=["mae"]),
        ak.ClassificationHead(
            loss="categorical_crossentropy", metrics=["accuracy"]
        ),
    ],
    overwrite=True,
    max_trials=2,
)
# Fit the model with prepared data.
model.fit(
    [image_data, image_data2],
    [classification_target, classification_target2],
    epochs=1,
    batch_size=3,
)
```

```python
model = model.export_model()
```

```python
keras.utils.plot_model(model, to_file="cikti.png", show_shapes=True)
```
